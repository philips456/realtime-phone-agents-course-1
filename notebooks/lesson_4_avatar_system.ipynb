Let's see how it works with our FastRTCAGent.

```python
from realtime_phone_agents.stt import get_stt_model
from realtime_phone_agents.agent.fastrtc_agent import FastRTCAgent
from realtime_phone_agents.tts import get_tts_model
```

```python
whisper = get_stt_model("whisper-groq")
orpheus = get_tts_model("together")
```

As default, we'll always use `tara`.

```python
FastRTCAgent(
    stt_model=whisper,
    tts_model=orpheus,
    thread_id="test"
).stream.ui.launch()
```

But now, thanks to the different voices, we can change the actual "agent" we are talking to!

```python
AVATAR_NAME = "leo"

orpheus.set_voice(AVATAR_NAME)

FastRTCAgent(
    stt_model=whisper,
    tts_model=orpheus,
    thread_id="test",
    avatar=AVATAR_NAME
).stream.ui.launch()
```

```python
AVATAR_NAME = "zoe"

orpheus.set_voice(AVATAR_NAME)

FastRTCAgent(
    stt_model=whisper,
    tts_model=orpheus,
    thread_id="test",
    avatar=AVATAR_NAME
).stream.ui.launch()
```
